# 2. Business understanding: problem definition and use case context

(from Emily's slides)

### **Claims fraud prevention**: 

Identifying patterns and spotting outliers to quickly identify claim fraud in real-time using ML and AI algorithms in the insurance industry is a crucial application for Fraud Prevention.

*Business value*: Reduce financial losses and operational costs ; Safeguards reputation while enhancing customer trust and regulatory compliance.

#### **Approach**

Develop and train ML models for real-time fraud detection. There are several approaches to consider:

    - Anomaly Detection: Utilize unsupervised learning algorithms to detect anomalies or outliers in the data. Claims that deviate significantly from the norm may be flagged as potential fraud.
    - Supervised Learning: Train a classification model on historical labelled data to predict whether a new claim is fraudulent or not.
    - Ensemble Models: Combine multiple models to improve accuracy and robustness.

#### **Typical Data Required**

Claim Information:
Claim data such as policyholder, amount and type.

Claim Documents:
Scanned or uploaded documents related to the claim, such as photos, police reports, medical records, etc.
System generated document metadata, for example uploaded datetime.

Historical Data:
Historical claims for the same policyholder and same type of insurance.

Investigation and Decision:
Fraud Indicator: indicating whether the claim is suspected of being fraudulent. 
Claim Decision: the final decision on the claim (approved, denied, pending, etc.)

Now we check correlations between numerical values

```python
def plot_correlation(data):
    '''
    plot correlation's matrix to explore dependency between features 
    '''
    # init figure size
    rcParams['figure.figsize'] = 15, 10
    fig = plt.figure()
    sns.heatmap(data.corr(), annot=True, fmt=".2f")
    plt.show()
    fig.savefig('corr.png')
    
plot_correlation(data)
```

Most related features with label are: vehicle_claim, total_claim amount, property_claim, injury_claim. These features have high correlation between them (makes some sense); these features are also correlated to incident_hour_of_the_day and number_of_vehicles_involved (these two also have some correlation between them); age is also very correlated to months as customer.

Now we check features densities depending on the outcome values

```python
def plot_densities(data, label, dataset_name):
    '''
    Plot features densities depending on the outcome values
    '''
    
    names = list(data.columns)
    data = data.astype(float)
    # change fig size to fit all subplots beautifully 
    rcParams['figure.figsize'] = 15, 30

    # separate data based on outcome values 
    outcome_0 = data[data[label] == 0]
    outcome_1 = data[data[label] == 1]

    # init figure
    fig, axs = plt.subplots(18, 1)
    fig.suptitle('Features densities for different outcomes 0/1')
    plt.subplots_adjust(left = 0.25, right = 0.9, bottom = 0.1, top = 0.95,
                        wspace = 0.2, hspace = 0.9)

    # plot densities for outcomes
    for column_name in names[:18]: 
        ax = axs[names.index(column_name)]
        #plt.subplot(4, 2, names.index(column_name) + 1)
        outcome_0[column_name].plot(kind='density', ax=ax, subplots=True, 
                                    sharex=False, color="red", legend=True,
                                    label=column_name + ' for Outcome = 0')
        outcome_1[column_name].plot(kind='density', ax=ax, subplots=True, 
                                     sharex=False, color="green", legend=True,
                                     label=column_name + ' for Outcome = 1')
        ax.set_xlabel(column_name + ' values')
        ax.set_title(column_name + ' density')
        ax.grid('on')
    plt.show()
    fig.savefig(dataset_name+'_densities.png')

```

# 5. Modeling

## 5.1 Insurance claims

```python
# split data 

data_mod=data.drop(['policy_bind_date', 'incident_date', 'policy_csl'], axis = 1) # drop dates

X, y = data_mod.drop(['fraud_reported'], axis=1),data_mod.fraud_reported

cat = X.select_dtypes(include=['object']).columns.tolist()
cat
X = pd.get_dummies(X, columns=cat)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=30, random_state=35)
```

### Classification models

```python

model = GaussianNB()
model_2 = neighbors.KNeighborsClassifier(n_neighbors=5)
model_3 = DecisionTreeClassifier()
model_4 = XGBClassifier()
```

```python
# Model training
# gaussian naive bayes
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

print('Accuracy score for NB:',accuracy_score(y_test, y_pred))
```

```python
# knn
model_2.fit(X_train, y_train)
y_pred_2 = model_2.predict(X_test)
print('Accuracy score for KNN:',accuracy_score(y_test, y_pred_2))
```

```python
# decision trees
model_3.fit(X_train, y_train)
y_pred_3 = model_3.predict(X_test)
print('Accuracy score for Trees:',accuracy_score(y_test, y_pred_3))
```

```python
# xgboost
model_4.fit(X_train, y_train)
y_pred_4 = model_4.predict(X_test)
print('Accuracy score for xgboost:',accuracy_score(y_test, y_pred_4))
```

```python
acc = pd.DataFrame()
acc['model'] = ['NB', 'KNN', 'Trees', 'XGB']
acc['accuracy score'] = [accuracy_score(y_test, y_pred),accuracy_score(y_test, y_pred_2),accuracy_score(y_test, y_pred_3),
                         accuracy_score(y_test, y_pred_4)]
acc
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred), classification_report(y_test, y_pred_2),classification_report(y_test, y_pred_3),
     classification_report(y_test, y_pred_4))
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test, y_pred), '\n',confusion_matrix(y_test, y_pred_2),'\n',confusion_matrix(y_test, y_pred_3),'\n',
      confusion_matrix(y_test, y_pred_4))
```

```python
print(model_4)
```

```python
feature_important = model_4.get_booster().get_score(importance_type='gain')
keys = list(feature_important.keys())
values = list(feature_important.values())
feature = pd.DataFrame(data=values, index=keys, columns=["score"]).sort_values(by = "score", ascending=False)
feature.nlargest(40, columns="score").plot(kind='barh', figsize = (20,10)) ## plot top 40 features
show()
```
